```python
from data import create_data
from model import HME

import matplotlib.pyplot as plt

import numpy as np
```

```python
import pickle

with open('hme_3class.pkl', 'rb') as file:
    hme3 = pickle.load(file)
    
with open('hme_2class.pkl', 'rb') as file:
    hme2 = pickle.load(file)
```

```python
n_feature = 2
n_sample = 2000
n_level = 2
x_2, y_2 = create_data(n_sample, n_feature, n_level)
batch_size = 2000
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=4)
rf.fit(x_2, y_2.ravel())
y_pred_rf_2 = rf.predict(x_2)

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=4)
dt.fit(x_2, y_2.ravel())
y_pred_dt_2 = dt.predict(x_2)

y_pred_hme_2 = hme2.predict(x=x_2)
```

```python
n_feature = 2
n_sample = 2000
n_level = 3
x_3, y_3 = create_data(n_sample, n_feature, n_level)
batch_size = 2000
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=4)
rf.fit(x_3, y_3.ravel())
y_pred_rf_3 = rf.predict(x_3)

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=4)
dt.fit(x_3, y_3.ravel())
y_pred_dt_3 = dt.predict(x_3)

y_pred_hme_3 = hme3.predict(x=x_3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(25, 10))

ax[0][0].scatter(x_2[:, 0], x_2[:, 1], c=y_2, label='Real')
ax[0][0].legend(loc='upper right')

# ax[0][1].scatter(x_2[:, 0], x_2[:, 1], c=y_pred_lr_2, label='Logistic Regression')
# ax[0][1].legend(loc='upper right')

ax[0][1].scatter(x_2[:, 0], x_2[:, 1], c=y_pred_dt_2, label='Decision Tree')
ax[0][1].legend(loc='upper right')

ax[0][2].scatter(x_2[:, 0], x_2[:, 1], c=y_pred_rf_2, label='Random Forest')
ax[0][2].legend(loc='upper right')

ax[0][3].scatter(x_2[:, 0], x_2[:, 1], c=y_pred_hme_2, label='sHME')
ax[0][3].legend(loc='upper right')

ax[1][0].scatter(x_3[:, 0], x_3[:, 1], c=y_3, label='Real')
ax[1][0].legend(loc='upper right')

# ax[1][1].scatter(x_3[:, 0], x_3[:, 1], c=y_pred_lr_3, label='Logistic Regression')
# ax[1][1].legend(loc='upper right')

ax[1][1].scatter(x_3[:, 0], x_3[:, 1], c=y_pred_dt_3, label='Decision Tree')
ax[1][1].legend(loc='upper right')

ax[1][2].scatter(x_3[:, 0], x_3[:, 1], c=y_pred_rf_3, label='Random Forest')
ax[1][2].legend(loc='upper right')

ax[1][3].scatter(x_3[:, 0], x_3[:, 1], c=y_pred_hme_3, label='sHME')
ax[1][3].legend(loc='upper right')

# plt.savefig('syn.png')
```
